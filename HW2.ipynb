```python
# Извлечение имени из полного имени
df['FirstName'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[1].strip())

# Самое популярное мужское имя
male_names = df[df['Sex'] == 'male']['FirstName']
most_common_male_name = male_names.mode()[0]

# Самое популярное женское имя
female_names = df[df['Sex'] == 'female']['FirstName']
most_common_female_name = female_names.mode()[0]

print(f"Самое популярное мужское имя: {most_common_male_name}")
print(f"Самое популярное женское имя: {most_common_female_name}")
```

```python
# Группировка по классу и полу, затем нахождение самого популярного имени
for pclass in df['Pclass'].unique():
    for sex in ['male', 'female']:
        names = df[(df['Pclass'] == pclass) & (df['Sex'] == sex)]['FirstName']
        most_common_name = names.mode()[0]
        print(f"Класс {pclass}, пол {sex}: {most_common_name}")
```
